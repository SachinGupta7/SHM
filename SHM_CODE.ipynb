# Energy expression


Okay so now we will find the expression for Dimensionless energy. since i have already have $l_0$ and $t_0$ as a natural scale of length and time respectively, so natural scale of Energy can be defined as 

\begin{equation*}
E_0 = \frac{1}{2}kl_0^2  = \frac{1}{2}mw_0^2l_0^2
\end{equation*}

defining this as a scale of energy for our system we can get an expression of dimensionless energy as 

\begin{align}
\tilde E = \frac{\tilde V ^2}{4\pi^2} +\tilde x^2  \   \ \   \  verify\ \  this !
\end{align}

just divide the total energy of oscillator by $E_0$ you will get $eq.(4)$





## <center> Energy Conservation Code </center>

###### I will be also plotting x(t) here using Matplotlib 

```python
import matplotlib.pyplot as plt  ##importing libraries
import numpy as np
from matplotlib import style
style.use('Solarize_Light2') ##matplotlib style
x = 0 ##initial position
v = 2  ##initial velocity
a = -4*np.pi*np.pi*x  ##initial acceleration 
t = 0  ##initial time 
Energy  =  (v*v)/(4*np.pi*np.pi) + x*x  ##initial total energy conserved
Energy_arr = [Energy]  #3energy list  
dt = 0.01  ##time step size
end_time = 4.0  ##end time  

X = [x]   #3position list
T = [t]  ##time list

while(t<=end_time):
    ##these four lines are verlet algorithm term
    x+= dt*v/2   ##updating position
    a = -4*np.pi*np.pi*x  ##updating
    v += a*dt ## updating velocity
    x+= dt*v/2 #3updating position again
    
    
    t += dt  ## updating time
    Energy  =  (v*v)/(4*np.pi*np.pi) + x*x  ## updating energy 
    
    Energy_arr.append(Energy)  ## storing them in a list
    
    X.append(x)
    T.append(t)
#############################333
##plotting datas
plt.figure(figsize = (12,9))
plt.plot(T,Energy_arr,label = "Total Energy") 
plt.plot(T,X,label  =   "Postion vs Time")
plt.legend()
plt.show()
```
