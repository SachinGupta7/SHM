# This will run only for $VPython$ Kernel

```python
from vpython import*
import numpy as np



RL  = 2 #relaxed_length

#making figures
canvas(background=color.white,eight=400,width=700)

spring = helix(pos = vector(0,RL,0),radius = 0.03,
                     axis = vector(0,-1,0),length = 2,coils=30,
                     color = color.blue,thickness = 0.01)
base = box(pos = vector(0,2,0),length=1,height=0.01,width=0.1,color = color.green)
block = box(pos = vector(0,0,0),radius = 0.5,color = color.red,length = 0.5,width = 0.5,height = 0.5)

graph(height=400,width=700)
f1 = gcurve(color = color.cyan)

x = 0  ###initial position
V = 1 ##velocity
a = -4*np.pi*np.pi*x ##acceleration
t = 0  ##time
dt = 0.01
end_time = 10

while (t<=end_time):
    rate(50)
    x += dt*V/2
    a = -4*np.pi*np.pi*x
    V += dt*a
    x+= V*dt/2
    t+=dt
    spring.length = RL-x 
    block.pos.y = x 
    f1.plot( t,x)
```

# Simple Python Code

```python
import matplotlib.pyplot as plt
import numpy as np
x = 1
v = 0
a = -4*np.pi*np.pi*x
t = 0
dt = 0.01
end_time = 4.0

X = [x]
T = [t]

while(t<=end_time):
    x+= dt*v/2
    a = -4*np.pi*np.pi*x
    v += a*dt
    x+= dt*v/2
    t += dt
    
    X.append(x)
    T.append(t)
plt.figure(figsize = (12,9))
plt.plot(T,X)    
plt.show()
```
